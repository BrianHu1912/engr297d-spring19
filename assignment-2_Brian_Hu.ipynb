```python
#Linear Regression
import numpy as np
import random
import math
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False 


true_w = np.array([3])
inp = []
output = []
d = len(true_w)
points = []

for i in range(10000):
    x = np.random.randn(d)
    y = true_w.dot(x)+np.random.randn()
    points.append((x,y))
    inp.append(x)
    output.append(y)

inp = np.asarray(inp)
plt.scatter(inp[:, 0], output, s=0.1, color='k', marker='o')
plt.plot([-3.3,3.3],[-10,10])
```

```python
def F(w):
    value = 0
    for x,y in points:
        value += ((x.dot(w)-y)**2)
    return value/len(points)
```

```python
def dF(w):
    value = 0
    for x,y in points:
        value += 2*(x.dot(w)-y)*x
    return value/len(points)
```

```python
def gradientD():
    w = np.zeros(1)
    l = .01
    for i in range(50):
        f = F(w)
        g = dF(w)
        w = w - l * g
        print("iteration : {},w : {}, f : {}, g : {}".format(i,w,f,g))
gradientD()
```
